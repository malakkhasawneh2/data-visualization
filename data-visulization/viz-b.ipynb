```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
sns.set()
sns.get_dataset_names()
```

```python

iris= sns.load_dataset('iris')
iris= pd.DataFrame(iris)
iris
```

```python
iris.head(20)
```

```python
sns.set_theme(style="white")
sns.relplot(x="sepal_length", y="sepal_width", hue="species", alpha=.5, palette="muted",height=6, data=iris)
```

```python
sns.displot(data=iris, x="sepal_length", hue="species", col="species")
```

# Results
* The setota species have more sepal width than others but also the sepal length is les than other species

* Versicolor and virginica they look similar but virginica have more sepals are higher at length
