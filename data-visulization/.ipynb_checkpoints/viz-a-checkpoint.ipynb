```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
sns.set()
sns.get_dataset_names()
```

```python

diamonds= sns.load_dataset('diamonds')
diamonds= pd.DataFrame(diamonds)
diamonds
```

```python
diamonds.head(20)
```

```python
sns.set_theme(style="white")
sns.relplot(x="carat", y="price", hue="clarity", alpha=.5, palette="muted",height=6, data=diamonds)
```

```python
sns.displot(data=diamonds, x="clarity", hue="color", col="color")
```
